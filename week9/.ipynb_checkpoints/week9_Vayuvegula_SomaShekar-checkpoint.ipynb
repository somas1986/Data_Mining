```python
#Import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df_loan = pd.read_csv("Loan_Train.csv")
```

```python
df_loan.head()
```

# Drop the column “Load_ID.”

```python
df_loan = df_loan.drop("Loan_ID",axis=1)
```

```python
df_loan
```

# Drop any rows with missing data

```python
df_loan = df_loan.dropna()
```

```python
df_loan
```

# Convert the categorical features into dummy variables.

```python
df_loan = pd.get_dummies(df_loan,
                     columns = df_loan.select_dtypes(include='object').keys())
```

```python
df_loan.head()
```

# Split the data into a training and test set, where the “Loan_Status” column is the target.

```python
x_data = df_loan.loc[:, ~df_loan.columns.isin(['Loan_Status_Y', 'Loan_Status_N'])].values
y_data = df_loan[['Loan_Status_Y']].values
```

```python
x_data
```

```python
y_data
```

```python
#Training and tes datasets split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data ,test_size = 0.2, shuffle=False)
```

# Create a pipeline with a min-max scaler and a KNN classifier

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import VarianceThreshold
```

```python
#Create a pipeline

pipe= Pipeline([
("scaler", MinMaxScaler()),
("classifier", KNeighborsClassifier())
])
```

# Fit a default KNN classifier to the data with this pipeline. Report the model accuracy on the test set.

```python
#Fitting Pipeline Model
pipe.fit(x_train,y_train)
```

```python
#score the knn model on the testing data
pipe.score(x_test,y_test)
```
