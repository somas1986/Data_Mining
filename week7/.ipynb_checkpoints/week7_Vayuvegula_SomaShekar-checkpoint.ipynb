```python
#Import libararies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

# 1. Import the housing data as a data frame and ensure that the data is loaded properly

```python
df_housing = pd.read_csv("train.csv")
```

```python
df_housing.head()
```

# 2. Drop the "Id" column and any features that are missing more than 40% of their values

```python
df_housing.drop('Id',axis=1,inplace=True)
```

```python
df_housing.head()
```

```python
df_housing.dtypes
```

```python
housing_null = df_housing.isnull().sum() / len(df_housing)
missing_features = housing_null[housing_null > 0.40].index
```

```python
missing_features
```

```python
df_housing.drop(missing_features, axis=1, inplace=True)
```

```python
df_housing.head()
```

# 3. For numerical columns, fill in any missing data with the median value.

```python
num_cols = df_housing.select_dtypes(include='number')
```

```python
num_cols
```

```python
df_housing.fillna(num_cols.median().iloc[0], inplace=True)
```

```python
df_housing.head()
```

# 4. For categorical columns, fill in any missing data with the most common value (mode)

```python
catCols = df_housing.select_dtypes(include='object')
```

```python
catCols
```

```python
df_housing.fillna(catCols.mode().iloc[0], inplace=True)
```

```python
df_housing.head()
```

# 5.Convert the categorical columns to dummy variables

```python
df_housing = pd.get_dummies(df_housing,
                     columns = df_housing.select_dtypes(include='object').keys())
```

```python
df_housing.head()
```

# 6. Split the data into a training and test set, where the SalePrice column is the target

```python
x_data = df_housing.loc[:, df_housing.columns != 'SalePrice'].values
y_data = df_housing['SalePrice'].values
```

```python
x_data
```

```python
y_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data ,test_size = 0.2, shuffle=False)
```

# 7. Run a linear regression and report the R2-value and RMSE on the test set.

```python
#Load required libraries
from sklearn.linear_model import LinearRegression
```

# 12. Take your original training features (from step 6) and apply a min-max scaler to them.

```python
#Import minmax scaler libraries
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline
```

```python
# split the data into training and test
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data ,test_size = 0.2, shuffle=False)
```

```python
# define min max scaler
minmax_scaler = MinMaxScaler()
```

```python
# apply min-max scaling
minmax_scaler = MinMaxScaler(feature_range=(-1, 1))
scaled_features = minmax_scaler.fit_transform(x_data)
scaled_features[:3]
```

```python
# create a ml model using pipeline
minMaxModel = make_pipeline(MinMaxScaler(), LinearRegression())
```

```python
minMaxModel.fit(x_train, y_train)
```
