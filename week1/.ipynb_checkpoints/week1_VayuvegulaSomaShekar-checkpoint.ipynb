```python
#Find the five publishers with the highest total global sales. 
#Note: You will need to calculate the total global sales for each publisher to do this.

dataframe1 = dataframe.groupby('Publisher')['Global_Sales'].sum()

dataframe1.sort_values(ascending=False).head()
```

```python
#Create a new column in the data frame that calculates the percentage of global sales from North America. 
#Display the first five rows of the new data frame.

dataframe['Percentage_Global_Sales'] = (dataframe['NA_Sales'] / dataframe['Global_Sales']) * 100
dataframe.head()
```

```python
#Find the number NaN entries (missing data values) in each column.

dataframe.isnull().sum(axis = 0)
```

```python
#Try to calculate the median user score of all the video games. 
#You will likely run into an error because some of the user score entries are a non-numerical string that cannot 
#be converted to a float. Find and replace this string with NaN and then calculate the median. 
#Then, replace all NaN entries in the user score column with the median value.
dataframe['User_Score'] = dataframe['User_Score'].replace("tbd", np.nan)
dataframe['User_Score'].median()
```
